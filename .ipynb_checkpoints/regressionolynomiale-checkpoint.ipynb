```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
def afficher_pol(LPol):
    L = LPol[:];
    L.reverse()
    L=  [round(e,2) for e in L]
    for i in range(len(L)):
        print("("+str(L[i])+"X^"+str(i)+")",end="+")


```

```python
def Evaluer_pol(LPol,x):
    s=0
    L=LPol[:]
    L.reverse()
    for i in range(len(LPol)):
        s +=L[i](x*i)
        return(s)
```

```python
df = pd.read_csv(r"C:\Users\pc\Desktop\S7\MachineLearningTps\dataset_polynomial3.csv")
```

```python
X = df.iloc[:,0]
Y = df.iloc[:,1]
```

```python
degre_polynome=2
p=np.poly1d(np.polyfit(X,Y,degre_polynome))
```

```python
LPol=list(p)
afficher_pol(LPol)
```

```python
les_x = np.linspace(min(X),max(X),100)
plt.scatter(X,Y)
plt.plot(les_x,p(les_x),c="r")
plt.show()
```
