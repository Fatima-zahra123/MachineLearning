```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv(r"C:\Users\pc\Desktop\S7\MachineLearningTps\images_chiffres_codees_niveau_de_gris.csv")
data
```

```python
type(data)
```

```python
#convert dataframe to matrix
matrix = data.to_numpy()
type(matrix)
matrix.shape
```

```python
#preparing dataset ==> isolate outputs and inputs
inputs= matrix[0:2100,1:]
outputs = matrix[0:2100,0]
```

```python
clf = DecisionTreeClassifier()
clf.fit(inputs,outputs)
```

```python
xtest = matrix[2100:,1:]
actual_label = matrix[2100:,0]
p= clf.predict(xtest)
p
```

```python
count =0
for i in range (0,459):
    if p[i]==actual_label[i]:
        count+=1
count
```

```python
print("Accuracy",(count/459)*100)
```
