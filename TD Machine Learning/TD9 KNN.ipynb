```python
from  collections import Counter
import math
def mean(labels):  #lables=target
    """calcul de la moyenne des k etiquettes"""
    return sum(labels)/len(labels)
def mode(labels): #C'est la (ou les) valeur(s) du caractère dont l'effectif est le plus grand.
    """calcul du mode des k etiquettes"""
    return Counter(labels).most_common(1)[0][0] #counter donne le nombre d'occurence
print(mode([0,1,2,2]))
def euclidean_distance(point1,point2):
    """Calcul de la distance euclidienne"""
    sum_squared_distance=0
    for i in range(len(point1)):
        sum_squared_distance+=math.pow(point1[i]-point2[i],2)
    return math.sqrt(sum_squared_distance)    
```

```python
def knn(data, query, k,distance_utilisee, choix_knn_regress_ou_classif): #query :valeur a predire
    """Algorithme des k plus proches voisins knn"""
    distances_et_positions = []
    for index, example in enumerate(data):
        distance=distance_utilisee(example[:-1],query) #chercher les voisins
        distances_et_positions.append((distance, index))
        #ici une liste de tuples sera triée selon les distances
        sorted_distances_et_positions=sorted(distances_et_positions) #on fait le tri
        k_proches_distances_and_indices=sorted_distances_et_positions[:k]
        k_proches_classes=[]
        for dist_indice in k_proches_distances_and_indices:
            num=dist_indice[1];k_proches_classes+=[data[num][1]]
            #7. si regression(choix_knn_regress_ou_classif = mean),
            #retourner la moyenne des k etiquettes
            #8. si classification, retourner le mode des k etiquettes
            return k_proches_distances_and_indices,\
        choix_knn_regress_ou_classif(k_proches_classes),k_proches_classes
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV,train_test_split
import pandas
import numpy as np
data=pandas.read_csv("C:\images_chiffres_codees_niveau_de_gris.csv")
data=np.array(data); x=data[ 0:1000,1:];y=data[0:1000,0];
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=0)
model=KNeighborsClassifier(n_neighbors=5, metric='manhattan')
model.fit(x_train,y_train); score=model.score(x_test,y_test)
print("score sur test-set par k=5 --> "+str(score*100)+'%')
distances =['euclidean','manhatten'];valeurs_de_k= np.arange(1,8)
parametres_grid={"n_neighbors":valeurs_de_k, 'metric':distances}
model=KNeighborsClassifier()
grid=GridSearchCV(model,parametres_grid,cv=3,scoring='accuracy')
grid.fit(x_train,y_train); print("best parametrs:",grid.best_params_)
model=grid.best_estimator_ ;model.fit(x_train,y_train)
score=model.score(x_test,y_test)
print("score sur test-set-parameteres automatiques="+str(score*100)+'%')
```
