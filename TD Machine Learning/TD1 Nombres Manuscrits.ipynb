```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from sklearn.tree import DecisionTreeClassifier

```

```python
#data=pandas.read_csv("C:\images_chiffres_codees_niveau_de_gris.csv").as_matrix()
data=pandas.read_csv("C:\images_chiffres_codees_niveau_de_gris.csv")
```

```python
data
#chaque ligne represente une image de taille(28*28)=784pixels
```

```python
dataMatrix=data.to_numpy() #convertir le tableau en matrice 
```

```python
dataMatrix
```

```python
clf=DecisionTreeClassifier()
x=dataMatrix[0:2100,1:]
label=dataMatrix[0:2100,0]
clf.fit(x,label) #entrainer le modele
xtest=dataMatrix[2100:,1:]
actual_label=dataMatrix[2100:,0]
p=clf.predict(xtext)
```

```python
p
```

```python
count=0
for i in range(0,459): #2559-2100=459 le nombre de xtest
    count+=1 if p[i]==actual_label[i]  else 0
print("Accuracy=",(count/459)*100)
    
```

```python
d=xtext[7]
Nombre_de_pixels_errones=30
for i in range(Nombre_de_pixels_errones):
    position=np.random.randint(0,784,1)[0]
    bruit=np.random.randint(-200,200,1)[0]
    d[position]+=bruit
    d[position]=d[position]%255
print(clf.predict([d]))
print("actual label=",actual_label[7])

```
