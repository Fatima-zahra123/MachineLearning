```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```python
def afficher_pol(LPol):
    L=LPol[:];L.reverse()
    L=[round(e,2)for e in L]
    for i in range(len(L)):
        print('('+str(L[i])+'*X^'+str(i)+')',end='+')
def Evaluer_pol(LPol,x):
    s=0
    L=LPol[:]
    L.reverse()
    for i in range(len(LPol)):
        s+=L[i]*(x**i)
    return(s)
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
data=pd.read_csv('C:\dataset_polynomial3.csv')
dataMatrix=data.to_numpy()
x=dataMatrix[1:1001,0]
y=dataMatrix[1:1001,1]
X_train, X_test, y_train, y_test = train_test_split(x,y, random_state=42,shuffle=True,test_size=0.3) #la methode holdout 
data
```

```python
degre_polynome=3
p=np.poly1d(np.polyfit(X_train,y_train,degre_polynome))
LPol=list(p)
afficher_pol(LPol)
les_x=np.linspace(min(X_train),max(X_train),100)
plt.scatter(X_train,y_train)
plt.plot(les_x,p(les_x),c='r')
plt.show()
#print("y_estimé(x_text) = ",Evaluer_pol(LPol,X_test))
#print("***********************************")
#print("y_vrai(x_text) = ",y_test)
```
