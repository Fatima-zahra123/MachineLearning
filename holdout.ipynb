```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
def afficher_pol(LPol):
    L = LPol[:];
    L.reverse()
    L=  [round(e,2) for e in L]
    for i in range(len(L)):
        print("("+str(L[i])+"X^"+str(i)+")",end="+")
```

```python
def Evaluer_pol(LPol,x):
    s=0
    L=LPol[:]
    L.reverse()
    for i in range(len(LPol)):
        s +=L[i]*(x^i)
    return(s)
```

```python
df = pd.read_csv(r"C:\Users\pc\Desktop\S7\MachineLearningTps\dataset_polynomial3.csv")
X = df.iloc[:,0].values
y = df.iloc[:,1].values
X
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33 ,shuffle=False)
```

```python
print(X_train,X_test,y_train , y_test)
```

```python
degre_polynome=3
p=np.poly1d(np.polyfit(X_train,y_train,degre_polynome))
```

```python
LPol=list(p)
afficher_pol(LPol)
```

```python
les_x = np.linspace(min(X_train),max(X_train),100)
plt.scatter(X_train,y_train)
plt.plot(les_x,p(les_x),c="r")
plt.show()
```
